```python
def createMatrix(n):
    matrix=[]
    for i in range(n):
        r=[]
        for j in range(n):
            r.append(0)
        matrix.append(r)
    return matrix

def printMatrix(m):
    print('[')
    for r in m:
        print(r)
    print(']')

f=open("g.txt",'r')
vertex_edges=f.readline()
v_e=vertex_edges.split()
vertex=int(v_e[0])
edge=int(v_e[1])

G=createMatrix(vertex)

for i in range(edge):
    u_v=f.readline()
    u_v_list=u_v.split()
    u=int(u_v_list[0])
    v=int(u_v_list[1])
    G[u][v]=1
    #M[v][u]=1

printMatrix(G)  

import queue
def dfs(G,s,vertex):
    stack=queue.LifoQueue()
    visited=[0]*vertex
    stack.put(s)
    visited[s]=1
    while not stack.empty():
        next=stack.get()
        print(next,end=" ")
      
        for i in range(len(G[0])):
            if(G[next][i]==1 and visited[i]==0):
                stack.put(i)
                visited[i]=1
                
    

dfs(G,0,vertex)
    
```

```python
def createMatrix(n):
    matrix=[]
    for i in range(n):
        r=[]
        for j in range(n):
            r.append(0)
        matrix.append(r)
    return matrix

def printMatrix(m):
    print('[')
    for r in m:
        print(r)
    print(']')

f=open("g.txt",'r')
vertex_edges=f.readline()
v_e=vertex_edges.split()
vertex=int(v_e[0])
edge=int(v_e[1])

G=createMatrix(vertex)

for i in range(edge):
    u_v=f.readline()
    u_v_list=u_v.split()
    u=int(u_v_list[0])
    v=int(u_v_list[1])
    G[u][v]=1
    #M[v][u]=1

printMatrix(G)  

import queue
def dfs(G,s,vertex):
    stack=queue.LifoQueue()
    exploredset=set()
    stack.put(s)
    exploredset.add(s)
    while not stack.empty():
        next=stack.get()
        print(next,end=" ")
      
        for i in range(len(G[0])):
            if(G[next][i]==1 and i not in exploredset):
                stack.put(i)
                exploredset.add(i)
                



dfs(G,0,vertex)
    
```
