```python
def createEmptyList(vertex):
    adj_list={}
    for i in range(vertex):
        adj_list[i]=[]
    return adj_list

def printAdjList(adj_list):
    for k,v in adj_list.items():
        print(str(k)+"->"+str(v))


f=open("ug.txt",'r')
vertex_edges=f.readline()
v_e=vertex_edges.split()
vertex=int(v_e[0])
edge=int(v_e[1])

Adj_list=createEmptyList(vertex)
printAdjList(Adj_list)


for i in range(edge):
    u_v=f.readline()
    u_v_list=u_v.split()
    u=int(u_v_list[0])
    v=int(u_v_list[1])
    w = int(u_v_list[2])  
    Adj_list[u].append((v, w))
    Adj_list[v].append((u, w)) 


print("\n")
printAdjList(Adj_list)
```

```python
def createEmptyList(vertex):
    adj_list={}
    for i in range(vertex):
        adj_list[i]=[]
    return adj_list

def printAdjList(adj_list):
    for k,v in adj_list.items():
        print(str(k)+"->"+str(v))

def FindInDegree(adj_list, v):
    count = 0
    for u in adj_list:
        for n, w in adj_list[u]:  
            if n == v:
                count += 1
    print(f"In Degree of {v} is: {count}")
    

    
def FindOutDegree(adj_list, v):
    print(f"Out Degree of {v} is: {len(adj_list[v])}")
   

    
f=open("dg.txt",'r')
vertex_edges=f.readline()
v_e=vertex_edges.split()
vertex=int(v_e[0])
edge=int(v_e[1])

Adj_list=createEmptyList(vertex)
printAdjList(Adj_list)


for i in range(edge):
    u_v=f.readline()
    u_v_list=u_v.split()
    u=int(u_v_list[0])
    v=int(u_v_list[1])
    w = int(u_v_list[2])  
    Adj_list[u].append((v, w))
    #Adj_list[v].append((u, w)) 


print("\n")
printAdjList(Adj_list)

FindOutDegree(Adj_list,2)
FindInDegree(Adj_list,2)
```
