```python
import random
def createEmptyBoard(n):
    matrix=[]
    for i in range(n):
        r=[]
        for j in range(n):
            r.append(0)
        matrix.append(r)
    return matrix

def printBoard(m):
    print('[')
    for r in m:
        print(r)
    print(']')


def placeQueen(b):
    n=len(A[0])
    
    for column in range(n):
        row=random.randint(0,n-1)
        b[row][column]=1

    return b

def chromosome(n):
    chrom=""
    for i in range(n):
        
        gene=random.randint(0,n-1)
        chrom+=str(gene)

    return chrom
    
def createpopulation(p,n1):
    
    l=[]
    for i in range(p):
        
        l.append(chromosome(n1))

    return l
    
    
def Findattack(chrom):
    Attack=0
    n=len(chrom)
    for i in range(0,n-1):
        for j in range(i+1,n):
            if chrom[i]==chrom[j]:
                Attack+= 1

            elif abs(int(chrom[i])-int(chrom[j]))==abs(i-j):
                Attack+=1

    return Attack
                
                
                
def FindNonAttackingPair(chrom):
    non_attack = 0
    n = len(chrom)
    for i in range(0, n-1):
        for j in range(i+1, n):
           
           if chrom[i] != chrom[j] and abs(int(chrom[i]) - int(chrom[j])) != abs(i - j):
                non_attack += 1
    return non_attack

def findfitness(population):
    fitness_dict = {}
    for chrom in population:
        attacks = Findattack(chrom)
        non_attacks = FindNonAttackingPair(chrom)
        fitness_dict[chrom] = {'attacks': attacks, 'non_attacks': non_attacks}
    return fitness_dict

    

n=8    
population_size=6
A=createEmptyBoard(n)
#printBoard(A)
bq=placeQueen(A)
printBoard(bq)
s_chrom=chromosome(n)
print(s_chrom)
population=createpopulation (population_size,n)
print(population)
#attacks = Findattack(s_chrom)
#print( attacks)
#nonattacks = FindNonAttackingPair(s_chrom)
#print(nonattacks)
fitness = findfitness(population)
print(fitness)
```
